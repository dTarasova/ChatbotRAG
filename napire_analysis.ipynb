```python
finance_df = df[df['Sector'] == 'Finance']
elicitation_techniques = finance_df[['Interviews', 'Analysis of existing documents', 'Risk analyses', 'Prototyping', 'Workshops and focus groups', '(Requirements) Reuse databases and guidelines', 'Design Thinking / Lean Startup', 'External experts', 'Observations', 'We do not elicit requirements (ourselves)']]
elicitation_techniques = elicitation_techniques.apply(lambda x: x.value_counts().get('Yes', 0))
elicitation_techniques = elicitation_techniques[elicitation_techniques > 0]
elicitation_techniques = elicitation_techniques.sort_values(ascending=False)
elicitation_techniques
```
